```python
import pandas as pd
import xml.etree.ElementTree as ET
import networkx as nx
import numpy as np
from sklearn.metrics.pairwise import euclidean_distances
import matplotlib.pyplot as plt
import json
import pickle
from google.colab import files  # Only for Colab

# Define column names for Alibaba dataset
server_event_cols = ["timestamp", "machineID", "event_type", "event_detail", "capacity_CPU", "capacity_memory", "capacity_disk"]
server_usage_cols = ["timestamp", "machineID", "util_CPU", "util_memory", "util_disk", "load1", "load5", "load15"]
batch_task_cols = ["create_timestamp", "modify_timestamp", "job_id", "task_id", "instance_num", "status", "plan_cpu", "plan_mem"]
batch_instance_cols = ["start_timestamp", "end_timestamp", "job_id", "task_id", "machineID", "status", "seq_no", "total_seq_no", "real_cpu_max", "real_cpu_avg", "real_mem_max", "real_mem_avg"]
container_event_cols = ["ts", "event", "instance_id", "machine_id", "plan_cpu", "plan_mem", "plan_disk", "cpuset"]
container_usage_cols = ["ts", "instance_id", "cpu_util", "mem_util", "disk_util", "load1", "load5", "load15", "avg_cpi", "avg_mpki", "max_cpi", "max_mpki"]

# Base path to Alibaba dataset (adjust as needed)
alibaba_path = "/content/drive/My Drive/FYP-LOGS/"

# Load Alibaba CSV files
df_server_event = pd.read_csv(alibaba_path + "server_event.csv", names=server_event_cols, header=None)
df_server_usage = pd.read_csv(alibaba_path + "server_usage.csv", names=server_usage_cols, header=None)
df_batch_task = pd.read_csv(alibaba_path + "batch_task.csv", names=batch_task_cols, header=None)
df_batch_instance = pd.read_csv(alibaba_path + "batch_instance.csv", names=batch_instance_cols, header=None)
df_container_event = pd.read_csv(alibaba_path + "container_event.csv", names=container_event_cols, header=None)
df_container_usage = pd.read_csv(alibaba_path + "container_usage.csv", names=container_usage_cols, header=None)

# Parse DAX workflow
def parse_dax(xml_file):
    tree = ET.parse(xml_file)
    root = tree.getroot()
    namespace = "{http://pegasus.isi.edu/schema/DAX}"
    G = nx.DiGraph()

    for job in root.findall(f"{namespace}job"):
        job_id = job.get("id")
        runtime = float(job.get("runtime", 1))
        G.add_node(job_id, runtime=runtime)

    for child in root.findall(f"{namespace}child"):
        child_id = child.get("ref")
        for parent in child.findall(f"{namespace}parent"):
            parent_id = parent.get("ref")
            G.add_edge(parent_id, child_id)

    return G

cybershake_dag = parse_dax("CyberShake_1000.xml")

# Process Alibaba data
df_batch_instance["execution_time"] = df_batch_instance["end_timestamp"] - df_batch_instance["start_timestamp"]
df_batch_instance = df_batch_instance[df_batch_instance["execution_time"] > 0]

task_execution_data = df_batch_instance.groupby("task_id").agg(
    execution_time=("execution_time", "mean"),
    cpu_usage=("real_cpu_avg", "mean"),
    memory_usage=("real_mem_avg", "mean"),
    data_size=("real_mem_max", "mean")  # Proxy for data transferred
).reset_index()

container_metrics = df_container_usage.groupby("instance_id").agg(
    cpu_util=("cpu_util", "mean"),
    mem_util=("mem_util", "mean")
).reset_index()
task_execution_data = task_execution_data.merge(
    container_metrics, left_on="task_id", right_on="instance_id", how="left"
)
task_execution_data["task_id"] = task_execution_data["task_id"].astype(str)

server_capacity = df_server_event.groupby("machineID").agg(
    cpu_capacity=("capacity_CPU", "mean"),
    memory_capacity=("capacity_memory", "mean")
).reset_index()

df_disk_usage = df_server_usage.groupby("machineID")["util_disk"].mean().reset_index()
df_disk_usage.rename(columns={"util_disk": "disk_usage"}, inplace=True)

# Improved task matching
def find_closest_task_multi_feature(task_id, task_execution_data, cybershake_dag):
    node_features = [
        cybershake_dag.nodes[task_id].get("runtime", 10),
        cybershake_dag.nodes[task_id].get("cpu_usage", 0.5)
    ]
    alibaba_features = task_execution_data[["execution_time", "cpu_usage"]].fillna(0.5).values
    distances = euclidean_distances([node_features], alibaba_features)[0]
    closest_idx = distances.argmin()
    return task_execution_data.iloc[closest_idx]

# Merge Alibaba data into DAG
for node in cybershake_dag.nodes():
    matching_task = find_closest_task_multi_feature(node, task_execution_data, cybershake_dag)
    cybershake_dag.nodes[node]["execution_time"] = matching_task["execution_time"]
    cybershake_dag.nodes[node]["cpu_usage"] = matching_task["cpu_usage"]
    cybershake_dag.nodes[node]["memory_usage"] = matching_task["memory_usage"] if not pd.isna(matching_task["memory_usage"]) else 0.5
    cybershake_dag.nodes[node]["cpu_util"] = matching_task["cpu_util"] if not pd.isna(matching_task["cpu_util"]) else 0.5
    cybershake_dag.nodes[node]["mem_util"] = matching_task["mem_util"] if not pd.isna(matching_task["mem_util"]) else 0.5
    cybershake_dag.nodes[node]["data_size"] = matching_task["data_size"] if not pd.isna(matching_task["data_size"]) else 10

# Simulate dynamic task arrivals
arrival_times = np.random.exponential(scale=10, size=len(cybershake_dag.nodes()))
for i, node in enumerate(cybershake_dag.nodes()):
    cybershake_dag.nodes[node]["arrival_time"] = arrival_times[i]

# Add disk usage and server capacity
for node in cybershake_dag.nodes():
    machine_id = df_batch_instance[df_batch_instance["task_id"] == node]["machineID"].iloc[0] if node in df_batch_instance["task_id"].values else None
    if machine_id and machine_id in df_disk_usage["machineID"].values:
        cybershake_dag.nodes[node]["disk_usage"] = df_disk_usage[df_disk_usage["machineID"] == machine_id]["disk_usage"].values[0]
    else:
        cybershake_dag.nodes[node]["disk_usage"] = 0.5

    if machine_id and machine_id in server_capacity["machineID"].values:
        cybershake_dag.nodes[node]["cpu_capacity"] = server_capacity[server_capacity["machineID"] == machine_id]["cpu_capacity"].values[0]
        cybershake_dag.nodes[node]["memory_capacity"] = server_capacity[server_capacity["machineID"] == machine_id]["memory_capacity"].values[0]
    else:
        cybershake_dag.nodes[node]["cpu_capacity"] = 100
        cybershake_dag.nodes[node]["memory_capacity"] = 128

# Power model
CLOUD_POWER_IDLE, CLOUD_POWER_MAX = 100, 400
EDGE_POWER_IDLE, EDGE_POWER_MAX = 3, 10

for node in cybershake_dag.nodes():
    cybershake_dag.nodes[node]["machine_type"] = "cloud" if np.random.rand() > 0.3 else "edge"
    cpu_usage = cybershake_dag.nodes[node]["cpu_usage"]
    if cybershake_dag.nodes[node]["machine_type"] == "cloud":
        power = CLOUD_POWER_IDLE + (CLOUD_POWER_MAX - CLOUD_POWER_IDLE) * cpu_usage
    else:
        power = EDGE_POWER_IDLE + (EDGE_POWER_MAX - EDGE_POWER_IDLE) * cpu_usage
    cybershake_dag.nodes[node]["power_usage"] = power

# Add edge attributes with variation
for parent, child in cybershake_dag.edges():
    # Base network parameters
    if cybershake_dag.nodes[parent]["machine_type"] != cybershake_dag.nodes[child]["machine_type"]:
        bandwidth_base, latency_base = 50, 20  # Edge-cloud link
        ccr_base = 0.025
    else:
        bandwidth_base, latency_base = 500, 5  # Same-tier link
        ccr_base = 0.005

    # Introduce variation
    bandwidth = bandwidth_base * np.random.uniform(0.8, 1.2)  # ±20% variation
    latency = latency_base * np.random.uniform(0.9, 1.1)      # ±10% variation
    data_size = float(cybershake_dag.nodes[parent]["data_size"]) * np.random.uniform(0.5, 1.5)  # ±50% variation based on parent task
    t_comm = data_size / bandwidth
    e_network = t_comm * 0.1  # Energy proportional to communication time
    ccr = ccr_base * (data_size / 10)  # Scale CCR with data size

    cybershake_dag.edges[parent, child].update({
        "CCR": ccr,
        "T_comm": t_comm,
        "E_network": e_network,
        "bandwidth": bandwidth,
        "network_latency": latency
    })

# Fill missing values
DEFAULTS = {"execution_time": 10, "cpu_usage": 0.5, "memory_usage": 0.5, "disk_usage": 0.5}
for node in cybershake_dag.nodes():
    for attr, default in DEFAULTS.items():
        if pd.isna(cybershake_dag.nodes[node].get(attr)):
            cybershake_dag.nodes[node][attr] = default

# Validate dataset
print("✅ Dataset generation complete!")
for node in list(cybershake_dag.nodes())[:5]:
    print(f"Task {node}: {cybershake_dag.nodes[node]}")
for parent, child in list(cybershake_dag.edges())[:5]:
    print(f"Edge {parent} → {child}: {cybershake_dag.edges[parent, child]}")

# Plot distributions
execution_times = [cybershake_dag.nodes[n]["execution_time"] for n in cybershake_dag.nodes()]
cpu_usages = [cybershake_dag.nodes[n]["cpu_usage"] for n in cybershake_dag.nodes()]
plt.hist(execution_times, bins=50, alpha=0.7, label="Execution Time (s)")
plt.hist(cpu_usages, bins=50, alpha=0.7, label="CPU Usage")
plt.legend()
plt.title("Dataset Attribute Distributions")
plt.show()

# Export dataset as JSON
dag_dict = {
    "nodes": [{"id": n, **cybershake_dag.nodes[n]} for n in cybershake_dag.nodes()],
    "edges": [{"source": u, "target": v, **cybershake_dag.edges[u, v]} for u, v in cybershake_dag.edges()]
}
with open("cybershake_dag.json", "w") as f:
    json.dump(dag_dict, f, indent=4)

# Export dataset as Pickle (faster loading)
with open("cybershake_dag.pkl", "wb") as f:
    pickle.dump(cybershake_dag, f)

# Auto-download in Colab (uncomment if running in Colab)
files.download("cybershake_dag.json")
files.download("cybershake_dag.pkl")

print("✅ Dataset saved as 'cybershake_dag.json' and 'cybershake_dag.pkl'")
```

```python
import pandas as pd
import xml.etree.ElementTree as ET
import networkx as nx
import json
import random
import numpy as np

# Define Alibaba dataset column names
server_event_cols = ["timestamp", "machineID", "event_type", "event_detail", "capacity_CPU", "capacity_memory", "capacity_disk"]
server_usage_cols = ["timestamp", "machineID", "util_CPU", "util_memory", "util_disk", "load1", "load5", "load15"]
batch_task_cols = ["create_timestamp", "modify_timestamp", "job_id", "task_id", "instance_num", "status", "plan_cpu", "plan_mem"]
batch_instance_cols = ["start_timestamp", "end_timestamp", "job_id", "task_id", "machineID", "status", "seq_no", "total_seq_no", "real_cpu_max", "real_cpu_avg", "real_mem_max", "real_mem_avg"]

# Load Alibaba data
alibaba_path = "/content/drive/My Drive/FYP-LOGS/"
df_server_event = pd.read_csv(alibaba_path + "server_event.csv", names=server_event_cols, header=None)
df_server_usage = pd.read_csv(alibaba_path + "server_usage.csv", names=server_usage_cols, header=None)
df_batch_task = pd.read_csv(alibaba_path + "batch_task.csv", names=batch_task_cols, header=None)
df_batch_instance = pd.read_csv(alibaba_path + "batch_instance.csv", names=batch_instance_cols, header=None)

# Compute execution time and filter invalid values
df_batch_instance["execution_time"] = df_batch_instance["end_timestamp"] - df_batch_instance["start_timestamp"]
df_batch_instance = df_batch_instance[df_batch_instance["execution_time"] > 0]

# Aggregate execution and resource usage per task
task_execution_data = df_batch_instance.groupby("task_id").agg(
    execution_time=("execution_time", "mean"),
    cpu_usage=("real_cpu_avg", "mean"),
    memory_usage=("real_mem_avg", "mean")
).reset_index()
task_execution_data["task_id"] = task_execution_data["task_id"].astype(str)
print(f"Loaded {len(task_execution_data)} tasks from Alibaba batch_instance")

# Parse CyberShake_1000.xml
def parse_cybershake(xml_file):
    tree = ET.parse(xml_file)
    root = tree.getroot()
    namespace = "{http://pegasus.isi.edu/schema/DAX}"
    G = nx.DiGraph()

    for job in root.findall(f"{namespace}job"):
        job_id = job.get("id")
        runtime = float(job.get("runtime", 1))  # Default 1s if missing
        G.add_node(job_id, runtime=runtime)

    for child in root.findall(f"{namespace}child"):
        child_id = child.get("ref")
        for parent in child.findall(f"{namespace}parent"):
            parent_id = parent.get("ref")
            if parent_id in G.nodes and child_id in G.nodes:
                G.add_edge(parent_id, child_id)

    return G

cybershake_dag = parse_cybershake("Montage_1000.xml")
print(f"Parsed {len(cybershake_dag.nodes())} nodes and {len(cybershake_dag.edges())} edges from CyberShake_1000.xml")

# Verify 1000 nodes
if len(cybershake_dag.nodes()) != 1000:
    raise ValueError(f"Expected 1000 nodes, got {len(cybershake_dag.nodes())}. Check your CyberShake_1000.xml file.")

# Match Alibaba data to CyberShake tasks
def find_closest_task(task_id, task_execution_data, threshold=2.0):
    node_runtime = cybershake_dag.nodes[task_id].get("runtime", 10)
    similar_tasks = task_execution_data[
        (task_execution_data["execution_time"] - node_runtime).abs() <= threshold
    ]
    return similar_tasks.sample(1).iloc[0] if not similar_tasks.empty else None

# Integrate disk usage from server_usage
df_disk_usage = df_server_usage.groupby("machineID")["util_disk"].mean().reset_index()
df_disk_usage.rename(columns={"util_disk": "disk_usage"}, inplace=True)

for node in cybershake_dag.nodes():
    matching_task = find_closest_task(node, task_execution_data)
    if matching_task is not None:
        cybershake_dag.nodes[node]["execution_time"] = matching_task["execution_time"]
        cybershake_dag.nodes[node]["cpu_usage"] = min(2.0, matching_task["cpu_usage"])
        cybershake_dag.nodes[node]["memory_usage"] = min(0.5, matching_task["memory_usage"])
    else:
        cybershake_dag.nodes[node]["execution_time"] = 10
        cybershake_dag.nodes[node]["cpu_usage"] = 0.5
        cybershake_dag.nodes[node]["memory_usage"] = 0.5

    # Add disk usage (random machineID mapping if not available)
    machine_id = random.choice(df_disk_usage["machineID"].values)
    disk_usage = df_disk_usage[df_disk_usage["machineID"] == machine_id]["disk_usage"].values[0] if machine_id in df_disk_usage["machineID"].values else 0.5
    cybershake_dag.nodes[node]["disk_usage"] = disk_usage

    # Add edge-cloud attributes
    machine_type = "edge" if random.random() < 0.06 else "cloud"  # 6% edge
    cybershake_dag.nodes[node]["machine_type"] = machine_type
    cybershake_dag.nodes[node]["power_usage"] = (3 + 7 * cybershake_dag.nodes[node]["cpu_usage"]) if machine_type == "edge" else (100 + 300 * cybershake_dag.nodes[node]["cpu_usage"])
    cybershake_dag.nodes[node]["network_latency"] = 30.0 if machine_type == "edge" else random.uniform(1.0, 5.0)
    cybershake_dag.nodes[node]["migration_energy"] = random.uniform(2.5, 6.0) if machine_type == "edge" else random.uniform(0.1, 0.3)
    cybershake_dag.nodes[node]["data_size"] = 10
    cybershake_dag.nodes[node]["CCR"] = 0.005 if machine_type == "cloud" else 0.025
    if machine_type == "edge":
        cybershake_dag.nodes[node]["edge_queue_delay"] = 7.1

print("Merged Alibaba data with CyberShake DAG")

# Handle missing values
for node in cybershake_dag.nodes():
    for attr in ["execution_time", "cpu_usage", "memory_usage", "disk_usage"]:
        if np.isnan(cybershake_dag.nodes[node][attr]):
            cybershake_dag.nodes[node][attr] = 0.5 if attr != "execution_time" else 10

# Assign edge attributes
for parent, child in cybershake_dag.edges():
    src_type = cybershake_dag.nodes[parent]["machine_type"]
    dst_type = cybershake_dag.nodes[child]["machine_type"]
    ccr = 0.025 if src_type == "edge" or dst_type == "edge" else 0.005
    t_comm = 0.2 if ccr > 0.01 else 0.02
    bandwidth = 50 if ccr > 0.01 else 500
    latency = 20 if ccr > 0.01 else 5
    cybershake_dag.edges[parent, child]["CCR"] = ccr
    cybershake_dag.edges[parent, child]["T_comm"] = t_comm
    cybershake_dag.edges[parent, child]["E_network"] = t_comm / 10
    cybershake_dag.edges[parent, child]["bandwidth"] = bandwidth
    cybershake_dag.edges[parent, child]["network_latency"] = latency

# Convert to JSON format
dag_json = {
    "nodes": {node: cybershake_dag.nodes[node] for node in cybershake_dag.nodes()},
    "edges": [[parent, child, cybershake_dag.edges[parent, child]] for parent, child in cybershake_dag.edges()]
}

# Save to file
with open("cybershake_dag.json", "w") as f:
    json.dump(dag_json, f, indent=4)

print(f"Saved DAG with {len(dag_json['nodes'])} nodes and {len(dag_json['edges'])} edges to cybershake_dag.json")
```

```python
import pandas as pd
import xml.etree.ElementTree as ET
import networkx as nx
import numpy as np
from sklearn.metrics.pairwise import euclidean_distances
import json
from google.colab import files

# Define Alibaba column names
server_event_cols = ["timestamp", "machineID", "event_type", "event_detail", "capacity_CPU", "capacity_memory", "capacity_disk"]
server_usage_cols = ["timestamp", "machineID", "util_CPU", "util_memory", "util_disk", "load1", "load5", "load15"]
batch_instance_cols = ["start_timestamp", "end_timestamp", "job_id", "task_id", "machineID", "status", "seq_no", "total_seq_no", "real_cpu_max", "real_cpu_avg", "real_mem_max", "real_mem_avg"]

# Base path to Alibaba data
alibaba_path = "/content/drive/My Drive/FYP-LOGS/"

# Load Alibaba data
df_server_event = pd.read_csv(alibaba_path + "server_event.csv", names=server_event_cols, header=None)
df_server_usage = pd.read_csv(alibaba_path + "server_usage.csv", names=server_usage_cols, header=None)
df_batch_instance = pd.read_csv(alibaba_path + "batch_instance.csv", names=batch_instance_cols, header=None)

# Parse DAX workflow
def parse_dax(xml_file):
    tree = ET.parse(xml_file)
    root = tree.getroot()
    namespace = "{http://pegasus.isi.edu/schema/DAX}"
    G = nx.DiGraph()
    for job in root.findall(f"{namespace}job"):
        job_id = job.get("id")
        runtime = float(job.get("runtime", 1))
        G.add_node(job_id, runtime=runtime)
    for child in root.findall(f"{namespace}child"):
        child_id = child.get("ref")
        for parent in child.findall(f"{namespace}parent"):
            parent_id = parent.get("ref")
            G.add_edge(parent_id, child_id)
    return G

# Process Alibaba data
df_batch_instance["execution_time"] = df_batch_instance["end_timestamp"] - df_batch_instance["start_timestamp"]
df_batch_instance = df_batch_instance[df_batch_instance["execution_time"] > 0]
task_execution_data = df_batch_instance.groupby("task_id").agg(
    execution_time=("execution_time", "mean"),
    cpu_usage=("real_cpu_avg", "mean"),
    memory_usage=("real_mem_avg", "mean"),
    data_size=("real_mem_max", "mean")
).reset_index()
task_execution_data["task_id"] = task_execution_data["task_id"].astype(str)
server_capacity = df_server_event.groupby("machineID").agg(
    cpu_capacity=("capacity_CPU", "mean"),
    memory_capacity=("capacity_memory", "mean")
).reset_index()

# Match tasks to Alibaba data
def find_closest_task(node_id, task_data, dag):
    node_features = [dag.nodes[node_id].get("runtime", 10), 0.5]
    alibaba_features = task_data[["execution_time", "cpu_usage"]].fillna(0.5).values
    distances = euclidean_distances([node_features], alibaba_features)[0]
    return task_data.iloc[distances.argmin()]

# Generate datasets
def generate_dataset(dax_file, output_file):
    dag = parse_dax(dax_file)

    # Node attributes
    for node in dag.nodes():
        match = find_closest_task(node, task_execution_data, dag)
        dag.nodes[node]["execution_time"] = match["execution_time"]
        dag.nodes[node]["cpu_usage"] = match["cpu_usage"]
        dag.nodes[node]["memory_usage"] = match["memory_usage"]
        dag.nodes[node]["data_size"] = match["data_size"] if not pd.isna(match["data_size"]) else 10
        dag.nodes[node]["arrival_time"] = np.random.exponential(scale=10)
        machine_id = df_batch_instance["machineID"].sample().iloc[0]
        if machine_id in server_capacity["machineID"].values:
            dag.nodes[node]["cpu_capacity"] = server_capacity[server_capacity["machineID"] == machine_id]["cpu_capacity"].iloc[0]
            dag.nodes[node]["memory_capacity"] = server_capacity[server_capacity["machineID"] == machine_id]["memory_capacity"].iloc[0]
        else:
            dag.nodes[node]["cpu_capacity"] = 100
            dag.nodes[node]["memory_capacity"] = 128
        dag.nodes[node]["machine_type"] = "cloud" if np.random.rand() > 0.3 else "edge"
        power = 100 + (400 - 100) * dag.nodes[node]["cpu_usage"] if dag.nodes[node]["machine_type"] == "cloud" else 3 + (10 - 3) * dag.nodes[node]["cpu_usage"]
        dag.nodes[node]["power_usage"] = power

    # Edge attributes with variation
    for parent, child in dag.edges():
        if dag.nodes[parent]["machine_type"] != dag.nodes[child]["machine_type"]:
            bandwidth_base, latency_base = 50, 20
            ccr_base = 0.025
        else:
            bandwidth_base, latency_base = 500, 5
            ccr_base = 0.005
        bandwidth = bandwidth_base * np.random.uniform(0.8, 1.2)
        latency = latency_base * np.random.uniform(0.9, 1.1)
        data_size = dag.nodes[parent]["data_size"] * np.random.uniform(0.5, 1.5)
        t_comm = data_size / bandwidth
        e_network = t_comm * 0.1
        ccr = ccr_base * (data_size / 10)
        dag.edges[parent, child].update({
            "CCR": ccr, "T_comm": t_comm, "E_network": e_network,
            "bandwidth": bandwidth, "network_latency": latency
        })

    # Save to JSON
    with open(output_file, "w") as f:
        json.dump(nx.node_link_data(dag), f, indent=4)
    print(f"Saved {output_file}")

# Generate datasets
generate_dataset("CyberShake_1000.xml", "cybershake_dag_1000.json")
generate_dataset("Montage_1000.xml", "montage_dag_1000.json")

# Download
files.download("/content/drive/My Drive/cybershake_dag.json")
files.download("/content/drive/My Drive/montage_dag.json")
```

```python
import numpy as np
def find_closest_memory_usage(task_node, cybershake_dag):
    """Finds the closest task based on CPU usage to estimate memory usage."""
    current_cpu = cybershake_dag.nodes[task_node]["cpu_usage"]
    closest_task = None
    min_diff = float("inf")

    for other_node in cybershake_dag.nodes():
        if other_node == task_node or np.isnan(cybershake_dag.nodes[other_node]["memory_usage"]):
            continue
        diff = abs(cybershake_dag.nodes[other_node]["cpu_usage"] - current_cpu)
        if diff < min_diff:
            min_diff = diff
            closest_task = other_node

    return cybershake_dag.nodes[closest_task]["memory_usage"] if closest_task else 0.5  # Default 0.5 GB


# Replace missing memory values with closest match
for node in cybershake_dag.nodes():
    if np.isnan(cybershake_dag.nodes[node]["memory_usage"]):
        closest_memory = find_closest_memory_usage(node, cybershake_dag)
        cybershake_dag.nodes[node]["memory_usage"] = closest_memory
        print(f"🔄 Assigned memory {closest_memory} from closest task to Task {node}")

print("✅ Missing memory values replaced using closest matching task!")
```

```python
# Define power model for Cloud & Edge
CLOUD_POWER_IDLE = 100  # W
CLOUD_POWER_MAX = 400  # W
EDGE_POWER_IDLE = 3  # W
EDGE_POWER_MAX = 10  # W

# Compute power per task based on CPU utilization
for node in cybershake_dag.nodes():
    cpu_usage = cybershake_dag.nodes[node]["cpu_usage"]
    if cybershake_dag.nodes[node].get("machine_type") == "cloud":
        power = CLOUD_POWER_IDLE + (CLOUD_POWER_MAX - CLOUD_POWER_IDLE) * cpu_usage
    else:  # Assume edge if not cloud
        power = EDGE_POWER_IDLE + (EDGE_POWER_MAX - EDGE_POWER_IDLE) * cpu_usage

    cybershake_dag.nodes[node]["power_usage"] = power

print("✅ Power consumption integrated (Edge & Cloud)!")
```

```python
# Define constants
EDGE_BANDWIDTH = 50  # MB/s (Edge network)
CLOUD_BANDWIDTH = 500  # MB/s (Cloud network)
P_TRANSMIT = 0.1  # Power (W) required for network transmission

for parent, child in cybershake_dag.edges():
    parent_task = cybershake_dag.nodes[parent]
    child_task = cybershake_dag.nodes[child]

    # Get data size (default = 10MB)
    data_size = parent_task.get("data_size", 10)

    # Assign bandwidth & latency based on machine type
    if parent_task["machine_type"] == "edge":
        bandwidth = EDGE_BANDWIDTH
        latency = 20  # Edge-to-cloud delay (ms)
    else:
        bandwidth = CLOUD_BANDWIDTH
        latency = 5  # Cloud-to-cloud delay (ms)

    # Compute communication time
    T_comm = data_size / bandwidth  # in seconds

    # Compute computation time of child task
    T_comp = child_task.get("execution_time", 1)  # Default = 1 sec

    # Compute CCR
    CCR = T_comm / T_comp

    # Compute energy consumption for data transfer
    E_network = P_TRANSMIT * T_comm  # Simple model

    # Store all edge attributes
    cybershake_dag.edges[parent, child].update({
        "CCR": CCR,
        "T_comm": T_comm,
        "E_network": E_network,
        "bandwidth": bandwidth,
        "network_latency": latency
    })

    print(f"🔹 Edge {parent} → {child}: {cybershake_dag.edges[parent, child]}")
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import LabelEncoder
import pandas as pd
from copy import deepcopy

# Load original data
with open('montage_dag.json') as f:
    data = json.load(f)

nodes = data['nodes']

# Create DataFrame
df = pd.DataFrame(nodes)

# Drop rows with NaN values (SMOTE can't handle NaNs)
df = df.dropna()

# Encode labels: cloud = 0, edge = 1
label_enc = LabelEncoder()
df['machine_type_encoded'] = label_enc.fit_transform(df['machine_type'])

# Separate features and target
X = df.drop(columns=['id', 'machine_type', 'machine_type_encoded'])
y = df['machine_type_encoded']

# Apply SMOTE
smote = SMOTE()
X_res, y_res = smote.fit_resample(X, y)

# Recreate node objects
resampled_nodes = []
for i, row in enumerate(X_res.to_dict(orient='records')):
    node = {**row}
    node['id'] = f"SMOTE_{i:05d}"
    node['machine_type'] = label_enc.inverse_transform([y_res[i]])[0]
    resampled_nodes.append(node)

# Save to new JSON
data['nodes'] = resampled_nodes
with open('balanced_smote_montage.json', 'w') as f:
    json.dump(data, f, indent=2)
```

```python
import json
import random
from copy import deepcopy

# Load the original dataset
with open('montage_dag.json', 'r') as f:
    data = json.load(f)

# Analyze original distribution
nodes = data['nodes']
cloud_tasks = [n for n in nodes if n['machine_type'] == 'cloud']
edge_tasks = [n for n in nodes if n['machine_type'] == 'edge']
print(f"Original: {len(cloud_tasks)} cloud tasks, {len(edge_tasks)} edge tasks")

# Target: 400 cloud and 400 edge tasks
TARGET_CLOUD = 400
TARGET_EDGE = 400
TOTAL_TARGET = TARGET_CLOUD + TARGET_EDGE

# Step 1: Balance the tasks
# Randomly select 400 cloud tasks and convert excess to edge
random.shuffle(cloud_tasks)
selected_cloud = cloud_tasks[:TARGET_CLOUD]
excess_cloud = cloud_tasks[TARGET_CLOUD:]

# Convert excess cloud tasks to edge-friendly tasks
for task in excess_cloud:
    task['machine_type'] = 'edge'
    # Reduce execution time for edge (faster, lower latency)
    task['execution_time'] = max(1.0, task['execution_time'] * 0.3)  # e.g., 10 -> 3
    # Lower power usage for edge (energy-efficient)
    task['power_usage'] = max(3.0, task['power_usage'] * 0.1)  # e.g., 268 -> 26.8
    # Adjust CPU/memory usage to fit edge constraints
    task['cpu_usage'] = min(task['cpu_usage'], 0.5)
    task['memory_usage'] = task['memory_usage'] if not isinstance(task['memory_usage'], float) else min(task['memory_usage'], 0.02)

# Combine edge tasks with converted tasks
all_edge = edge_tasks + excess_cloud
random.shuffle(all_edge)
selected_edge = all_edge[:TARGET_EDGE]

# New node list
new_nodes = selected_cloud + selected_edge
print(f"New: {len(selected_cloud)} cloud tasks, {len(selected_edge)} edge tasks")

# Step 2: Add monetary cost and adjust attributes
for node in new_nodes:
    # Add monetary cost (cloud: higher, edge: lower)
    if node['machine_type'] == 'cloud':
        node['monetary_cost'] = node['execution_time'] * 0.15  # $0.15/unit time
    else:
        node['monetary_cost'] = node['execution_time'] * 0.02  # $0.02/unit time
    # Ensure NaN memory_usage is handled
    if isinstance(node['memory_usage'], float) and not node['memory_usage'] > 0:
        node['memory_usage'] = 0.01  # Default small value

# Step 3: Adjust links to match new nodes
valid_ids = set(node['id'] for node in new_nodes)
new_links = [link for link in data['links'] if link['source'] in valid_ids and link['target'] in valid_ids]

# Step 4: Update the DAG
new_data = {
    "directed": True,
    "multigraph": False,
    "graph": {},
    "nodes": new_nodes,
    "links": new_links
}

# Save the new dataset
with open('balanced_montage_dag.json', 'w') as f:
    json.dump(new_data, f, indent=4)

print(f"New dataset saved as 'balanced_cybershake_dag.json' with {len(new_nodes)} nodes and {len(new_links)} links")
```
